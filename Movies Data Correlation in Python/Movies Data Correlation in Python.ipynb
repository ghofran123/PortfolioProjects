```python
# Import librairies

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Read the data
df = pd.read_csv(r'movies.csv')
```

```python
# Look at the data

df.head()
```

```python
# Check for missing data

for col in df.columns:
    pct_missing_Vals= np.mean(df[col].isnull())
    print('{} - {}%'.format(col,round(pct_missing_Vals*100)))
```

```python
# Change the data type for budget and gross and change the null value to 0

df['budget'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0).astype(int)
df['gross'] = pd.to_numeric(df['gross'], errors='coerce').fillna(0).astype(int)
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
pd.set_option('display.max_rows',None)
```

```python
df.sort_values(by=['gross'],inplace=False,ascending=False).head()
```

```python
# Drop any duplicates

df.drop_duplicates().head()
```

```python
# Scatter plot with Budget VS gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title(' Budget Vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
# Plot Budget vs Gross earning using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Looking at correlation in numeric values
# Pearson, Kendall, Spearman

numeric_df = df.select_dtypes(include=['number'])
numeric_df.corr(method='pearson')
```

```python
# Noticed High correlation between budget and gross (0.75)
```

```python
correlation_matrix = numeric_df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Budget for Film')
plt.show()
```

```python
# Look at Companies

df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
df.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Budget for Film')
plt.show()
```

```python
df_numerized.corr()
```

```python
# Create a new dataframe for correlations values

correlation_mat=df_numerized.corr()

# Display correlations by pairs

Corr_pairs = correlation_mat.unstack()
Corr_pairs
```

```python
# Sort correlation pairs by value

sorted_pairs = Corr_pairs.sort_values()
sorted_pairs
```

```python
# Extract high correlations (higher than 0.5)

high_corr = sorted_pairs[(sorted_pairs)>0.5]
high_corr
```

**We find strong positive correlations between movie budget and gross revenue, as well as between the number of votes and gross revenue. This suggests that higher budgets and greater audience engagement, measured by votes, tend to lead to higher gross revenues. These correlations highlight the importance of budget allocation and audience engagement in predicting a movie's financial success.**
